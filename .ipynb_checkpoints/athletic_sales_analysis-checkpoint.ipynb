```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_2020 = pd.read_csv('athletic_sales_2020.csv')
df_2021 = pd.read_csv('athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
df_2020.head()
```

```python
# Display the 2021 sales DataFrame
df_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
# Check the columns in the 2021 sales DataFrame
df_2020.info()
```

```python
# Check the 2021 sales data types.
df_2021.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df_combined = pd.concat([df_2020, df_2021], ignore_index=True)

# Display the first few rows of the combined DataFrame to verify
df_combined.head()
```

```python
# Check if any values are null.
if df_combined.isnull().sum().sum() == 0:
    print("No data is null.")
else:
    print("There are missing values in the dataset.")

# Display the null check table
df_combined.isnull().sum()
```

```python
# Check the data type of each column
df_combined.dtypes
```

```python
# Convert the 'invoice_date' to a datetime datatype by inferring the format
df_combined['invoice_date'] = pd.to_datetime(df_combined['invoice_date'], errors='coerce')

# Confirm the data type change
df_combined.dtypes
```

```python
# Confirm that the 'invoice_date' data type has been changed.
df_combined['invoice_date'] = pd.to_datetime(df_combined['invoice_date'], format='%Y-%m-%d')

# Check the data types of all columns
print("Updated data types:\n", df_combined.dtypes)

# Show the specific column that was changed
if df_combined['invoice_date'].dtype == 'datetime64[ns]':
    print("\nThe 'invoice_date' column has been successfully converted to datetime.")
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
df_pivot = pd.pivot_table(df_combined, values='units_sold', index=['region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold".
df_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
df_pivot = df_pivot.sort_values(by='Total_Products_Sold', ascending=False).head(5)

# Display the result
df_pivot
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
df_grouped_sales = df_combined.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the "total_sales" column to "Total Sales".
df_grouped_sales.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.
df_grouped_sales = df_grouped_sales.sort_values(by='Total_Sales', ascending=False).head(5)

# Display the result
df_grouped_sales

```

```python
# Show the total sales for the products sold for each region, state, and city.
df_pivot_sales = pd.pivot_table(df_combined, values='total_sales', index=['region', 'state', 'city'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales".
df_pivot_sales.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.
df_pivot_sales = df_pivot_sales.sort_values(by='Total_Sales', ascending=False).head(5)

# Display the result
df_pivot_sales
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_df = df_combined[df_combined['product'] == "Women's Athletic Footwear"]

# Display the first few rows to confirm the filter worked as expected
womens_footwear_df.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
df_womens_footwear_units_sold = womens_footwear_df.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold".
df_womens_footwear_units_sold.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
df_womens_footwear_units_sold = df_womens_footwear_units_sold.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)

# Display the result
df_womens_footwear_units_sold
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
df_womens_footwear_pivot = pd.pivot_table(womens_footwear_df, values='units_sold', index=['retailer', 'region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold".
df_womens_footwear_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
df_womens_footwear_pivot = df_womens_footwear_pivot.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)

# Display the result
df_womens_footwear_pivot
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column as the index, and the "total_sales" as the values.
df_womens_footwear_sales_by_day = pd.pivot_table(womens_footwear_df, values='total_sales', index='invoice_date', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales".
df_womens_footwear_sales_by_day.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
df_womens_footwear_sales_by_day.sort_values(by='Total Sales', ascending=False)
```

```python
# Resample the pivot table into daily bins and get the total sales for each day.
df_womens_footwear_sales_by_day_resampled = df_womens_footwear_sales_by_day.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
df_womens_footwear_sales_by_day_resampled_sorted = df_womens_footwear_sales_by_day_resampled.sort_values(by='Total Sales', ascending=False)

# Show the result
df_womens_footwear_sales_by_day_resampled_sorted.head()

```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
df_womens_footwear_sales_weekly = df_womens_footwear_sales_by_day.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
df_womens_footwear_sales_weekly_sorted = df_womens_footwear_sales_weekly.sort_values(by='Total Sales', ascending=False)

# Show the result
df_womens_footwear_sales_weekly_sorted.head()
```
